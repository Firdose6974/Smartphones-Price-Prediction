```python
def plot_graphs(column_name):
    sns.displot(data=df, x = column_name, kde=True, kind='hist', label=column_name)
    sns.catplot(data=df, x=column_name, kind='box')
    sns.countplot(data=df, x=column_name)
```

```python
for col in numerical_columns:
    plot_graphs(col)
```

# knn imputer
- It works only on numerical columns
- It takes each data point and cross check with other features and takes the closet point and we fill null vlaues with those close values

```python
# lets extract numerical columns
num_col = df.select_dtypes(include=['float64', 'int64']).drop(columns='price')
num_col.shape
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
num_col_values = imputer.fit_transform(num_col)
# lets convert the numpy array to dataframe
num_col_df = pd.DataFrame(data = num_col_values, columns = num_col.columns)
num_col_df
```

```python
num_col_df.isnull().sum()
```

```python
num_col_df['price'] = df['price']
num_col_df.head()
```

```python
# lets compare correlation with missing values and without missing values
```

```python
a = num_col_df.corr()['price'].reset_index()
```

```python
b = df.corr()['price'].reset_index()
```

```python
b.merge(a, on='index')
```
